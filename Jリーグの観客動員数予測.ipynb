```python
# home、away、stadium列を結合して重複を削除
unique_data = train[["home", "away", "stadium"]].drop_duplicates()

# 結果の表示
print(unique_data)
```

```python
train[['y']].info()
```

```python
unique_data[['stadium']].head()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np


# データ準備
home_data = unique_data[['home']]
away_data = unique_data[['away']]
stadium_data = unique_data[['stadium']]

# ワンホットエンコーディング
encoder = OneHotEncoder(handle_unknown="ignore")
home_encoded = encoder.fit_transform(np.array(home_data).reshape(-1, 1))
away_encoded = encoder.fit_transform(np.array(away_data).reshape(-1, 1))
stadium_encoded = encoder.fit_transform(np.array(stadium_data).reshape(-1, 1))

# 重回帰分析への取り込み
X = np.hstack([home_encoded.toarray(), away_encoded.toarray(), stadium_encoded.toarray()])
y = train[['y']] # 目的変数

# 重回帰分析の実行
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
unique_data[['stadium']].info()
```

```python
import pandas as pd

# データをDataFrameに変換
home_df = pd.DataFrame(home_data)
away_df = pd.DataFrame(away_data)
stadium_df = pd.DataFrame(stadium_data)
y_df = pd.DataFrame(y)

# データを結合
X = pd.concat([home_df, away_df, stadium_df], axis=1)
y = y_df

# ワンホットエンコーディング
encoder = OneHotEncoder(handle_unknown="ignore")
home_encoded = encoder.fit_transform(X[['home']].values)
away_encoded = encoder.fit_transform(X[['away']].values)
stadium_encoded = encoder.fit_transform(X[['stadium']].values)

# 重回帰分析への取り込み
X = np.hstack([home_encoded.toarray(), away_encoded.toarray(), stadium_encoded.toarray()])
y = np.array(y)

# 重回帰分析の実行
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
all = pd.concat([train, test])
```

```python
df = all[['home', 'away', 'stadium']]
```

```python
df = pd.get_dummies(df, dtype=np.int8)
df
```

```python
# train test 分割train分の長さで分割していけばきれいに分かれる
df_train = df[:len(train)]
df_test = df[len(train):]
```

# 重回帰分析

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

X = df_train

# 目的変数の作成
y = train[["y"]] # 目的変数

# 重回帰分析の実行
model = LinearRegression()
model.fit(X, y)
```

##yoshokuryo

```python
# 切片と係数の取得
print(model.intercept_)
print(model.coef_)

# R^2スコアの計算
from sklearn.metrics import r2_score
y_pred = model.predict(X)
print(r2_score(y, y_pred))
```

```python
# 予測される販売数
y_pred = model.predict(df_test)
print(y_pred)
```

# 提出

```python
sample_submit[1] = y_pred
sample_submit
```
